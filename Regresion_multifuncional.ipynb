```python
#Para hacer loops entre archivos
#listing es un arreglo de strings con los nombres de todos los archivos de la carpeta que tiene direccion: "path"

import os

file = open("Theta_Phi_Yo_Vo_g.txt", "w")

fallas_omitidas = 1000

path = dir_descargas+'/ComputationalMethodsData/homework/hw_4/Brahe-3141-f'

listing = os.listdir(path)
for infile in listing:
    if infile.endswith(".dat"): 
        archivo = genfromtxt(path+"/"+infile ,delimiter=" ")
        t = archivo[:,0]
        y = archivo[:,1]

        #Creamos las matrices A y A transpuesta a partir de los datos del tiempo
        At = zeros((3,38))
        A = zeros((38,3))
        for i in range(38):
            for j in range(3):            
                A[i,j] = t[i]**j
                At[j,i] = t[i]**j
                if j == 2:
                    A[i,j] = A[i,j]/2
                    At[j,i] = At[j,i]/2

        #Creamos la matriz Atranspuesta*A y a su ves el vector bprime que es Atranspuesta*y=bprime
        AtxA = zeros((3,3))
        bprime = zeros(3)
        for i in range (3):
            for j in range(3):
                for n in range(38):
                    AtxA[i,j] += At[i,n]*A[n,j]
                    
                    
        for i in range(3):
            for n in range(38):
                bprime[i] += At[i,n]*y[n]

        #Creamos constantes y las matrices triangular superior e inferior L y Ltranspuesta, donde L*Lt = At*A (descomposicion de Cholesky)
        a = 0.0
        s = 0.0
        L = zeros((3,3))
        Lt = zeros((3,3))

        for i in range(3):
            for j in range(i+1): 
                if i==j:
                    for n in range(i):        
                        a += (L[i,n])**2  
                    L[i,i] = (AtxA[i,i] - a)**0.5
                    a = 0.0
                else:
                    for n in range(j):    
                        s += L[j,n]*L[i,n]  
                    L[i,j] = (AtxA[i,j] - s)/L[j,j]
                    s = 0.0
            
        for i in range(3):
            for j in range(3):
                Lt[j,i] = L[i,j]

        #Creamos el vector z, donde L*(z)=bprime y resolvemos el sistema        
        z = zeros(3)       
        for i in range(3):
            k = 0.0
            for j in range(i):
                k += z[j]*L[i,j]
            z[i] = (bprime[i] - k)/L[i,i]

        #Resolvemos el sistema para el vector x, donde Lt*x = z
        x = zeros(3)
        h = 0.0

        for i in range(3):
            for j in range(3-i,3):
                h += x[j]*Lt[2-i,j]
            x[2-i] = (z[2-i]-h)/Lt[2-i,2-i]
            h = 0.0
        
        #Imprimimos el angulo Theta  Phi  y0  v0  g0, para los archivos no fallidos, y contamos cuantos fueron pasados por alto (fallidos)
        if x[0]<2:
            abc = infile
            arr = abc.split("_")
            linea = str(arr[2])+"\t"+str(arr[4].split(".dat")[0])+"\t"+str(x[0])+"\t"+str(x[1])+"\t"+str(x[2])+"\n"
            file.write(linea)
            fallas_omitidas = fallas_omitidas - 1

file.close()
print "El numero de archivos corruptos y omitidos fue " + str(fallas_omitidas)
```

```python
rcParams.update({'font.size': 15, 'font.family': 'sans-serif'})

archivo = genfromtxt(dir_descargas + "/Theta_Phi_Yo_Vo_g.txt",delimiter="\t")
theta = archivo[:,0]
g = -1*archivo[:,4]

figure(figsize=(8,8))
plot(theta, g, "o")
title(r'Magnitud de la Gravedad vs Theta',fontsize=18)
ylabel(r'Valor absoluto de la Gravedad($m/s^2$)', fontsize=15)
xlabel(r'Theta($grados$)', fontsize=15)
show()
```

```python
a = -1*g
gprom = mean(a)
F = 1 - a/gprom

figure(figsize=(8,8))
plot(theta, F, "o")
title(r'$gprom =$'+'$'+str(gprom)+'$\n'+' Desviacion de la gravedad del promedio vs Theta',fontsize=16)
ylabel(r'Magnitud de la deviacion', fontsize=15)
xlabel(r'Theta($grados$)', fontsize=15)
yticks([-0.15, -0.10, -0.05, 0.00, 0.05, 0.10])
show()
```

```python
# En el siguiente procedimiento hallaremos ao y a1 por medio de minimos cuadrados, valores del sistema de ecuaciones formado por:
# F[i] = ao*cos(2*pi*theta[i]/180) + a1 -----------> para todo i e {o, ... ,883}

#Creamos las matrices A y A transpuesta a partir de los datos de theta
At = zeros((2,884))
A = zeros((884,2))
for i in range(884):
    for j in range(2):
        if j ==0:
            A[i,j] = cos(2*pi*theta[i]/180)
            At[j,i] = cos(2*pi*theta[i]/180)
        else:
            A[i,j] = 1
            At[j,i] = 1

#Creamos la matriz Atranspuesta*A y a su ves el vector bprime que es Atranspuesta*F=bprime
AtxA = zeros((2,2))
bprime = zeros(2)
for i in range (2):
    for j in range(2):
        for n in range(884):
            AtxA[i,j] += At[i,n]*A[n,j]
            
            
for i in range(2):
    for n in range(884):
        bprime[i] += At[i,n]*F[n]

#Creamos constantes y las matrices triangular superior e inferior L y Ltranspuesta, donde L*Lt = At*A (descomposicion de Cholesky)
a = 0.0
s = 0.0
L = zeros((2,2))
Lt = zeros((2,2))

for i in range(2):
    for j in range(i+1): 
        if i==j:
            for n in range(i):        
                a += (L[i,n])**2  
            L[i,i] = (AtxA[i,i] - a)**0.5
            a = 0.0
        else:
            for n in range(j):    
                s += L[j,n]*L[i,n]  
            L[i,j] = (AtxA[i,j] - s)/L[j,j]
            s = 0.0
            
for i in range(2):
    for j in range(2):
        Lt[j,i] = L[i,j]

#Creamos el vector z, donde L*(Lt*x)=z y resolvemos el sistema        
z = zeros(2)       
for i in range(2):
    k = 0.0
    for j in range(i):
        k += z[j]*L[i,j]
    z[i] = (bprime[i] - k)/L[i,i]

#Resolvemos el sistema para el vector x
x = zeros(2)
h = 0.0

for i in range(2):
    for j in range(1-i,1):
        h += x[j]*Lt[1-i,j]
    x[1-i] = (z[1-i]-h)/Lt[1-i,1-i]
    h = 0.0

"""  
print A
print "--------------\n"
print At
print "--------------\n"
print AtxA
print "--------------\n"
print L
print "--------------\n"
print Lt
print "--------------\n"
print z
print "--------------\n"
print x
"""
```

```python
dispersion_coseno = zeros(884)
q = mean(F)

for i in range(884):
    dispersion_coseno[i] = 1 - (x[0]*cos(2*pi*theta[i]/180) + x[1])/q #F[i]
    
figure(figsize=(8,8))
plot(theta, dispersion_coseno, "o")
#title(r'Magnitud de la Gravedad vs Theta',fontsize=18)
#ylabel(r'Valor absoluto de la Gravedad($m/s^2$)', fontsize=15)
#xlabel(r'Theta($grados$)', fontsize=15)
show()
```

```python
# En el siguiente procedimiento hallaremos ao, a1, a2 por medio de minimos cuadrados, valores del sistema de ecuaciones formado por:
# F[i] = ao + a1*theta[i] + a2*theta[i]**2 -----------> para todo i e {o, ... ,883}

#Creamos las matrices A y A transpuesta a partir de los datos del tiempo
At = zeros((3,884))
A = zeros((884,3))
for i in range(884):
    for j in range(3):            
        A[i,j] = theta[i]**j
        At[j,i] = theta[i]**j
        
#Creamos la matriz Atranspuesta*A y a su ves el vector bprime que es Atranspuesta*y=bprime
AtxA = zeros((3,3))
bprime = zeros(3)
for i in range (3):
    for j in range(3):
        for n in range(884):
            AtxA[i,j] += At[i,n]*A[n,j]
                    
                    
for i in range(3):
    for n in range(884):
        bprime[i] += At[i,n]*F[n]

#Creamos constantes y las matrices triangular superior e inferior L y Ltranspuesta, donde L*Lt = At*A (descomposicion de Cholesky)
a = 0.0
s = 0.0
L = zeros((3,3))
Lt = zeros((3,3))

for i in range(3):
    for j in range(i+1): 
        if i==j:
            for n in range(i):        
                a += (L[i,n])**2  
            L[i,i] = (AtxA[i,i] - a)**0.5
            a = 0.0
        else:
            for n in range(j):    
                s += L[j,n]*L[i,n]  
            L[i,j] = (AtxA[i,j] - s)/L[j,j]
            s = 0.0
            
for i in range(3):
    for j in range(3):
        Lt[j,i] = L[i,j]

#Creamos el vector z, donde L*(z)=bprime y resolvemos el sistema        
z = zeros(3)       
for i in range(3):
    k = 0.0
    for j in range(i):
        k += z[j]*L[i,j]
    z[i] = (bprime[i] - k)/L[i,i]

#Resolvemos el sistema para el vector x, donde Lt*x = z
x = zeros(3)
h = 0.0

for i in range(3):
    for j in range(3-i,3):
        h += x[j]*Lt[2-i,j]
    x[2-i] = (z[2-i]-h)/Lt[2-i,2-i]
    h = 0.0
      
"""
print A
print "--------------\n"
print At
print "--------------\n"
print AtxA
print "--------------\n"
print L
print "--------------\n"
print Lt
print "--------------\n"
print z
print "--------------\n"
print x
"""
```

```python
dispersion_polinomio = zeros(884)
for i in range(884):
    dispersion_polinomio[i] = 1 - (x[0] + x[1]*theta[i] + x[2]*theta[i]**2)/q #F[i]
    
figure(figsize=(8,8))
plot(theta, dispersion_polinomio, "o")
#title(r'Magnitud de la Gravedad vs Theta',fontsize=18)
#ylabel(r'Valor absoluto de la Gravedad($m/s^2$)', fontsize=15)
#xlabel(r'Theta($grados$)', fontsize=15)
show()
```
